# **Seasonal and Weather Impact**

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Define the seasons based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create a 'Season' column
df['Season'] = df['Month'].apply(get_season)

# Group by 'Season' and 'Primary Type' to get counts
seasonal_crime_counts = df.groupby(['Season', 'Primary Type']).size().unstack()

# Plot the seasonal trends for each crime type
seasonal_crime_counts.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title('Seasonal Trends in Crime Types')
plt.xlabel('Season')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Creating our explicit dataset
df1 = df.groupby(['Month','Day','District','Hour'], as_index=False).agg({"Primary Type":"count"})
df1 = df1.sort_values(by=['District'], ascending=False)
df1.head()
```

```python
df1 = df1[['Month','Day','Hour','Primary Type','District']]
df1.head()
df1.shape
```

```python
# Feature Engineer and create a new feature
def crime_rate_assign(x):
    if(x<=7):
        return 0
    elif(x>7 and x<=15):
        return 1
    else:
        return 2
df1['Alarm'] = df1['Primary Type'].apply(crime_rate_assign)
df1 = df1[['Month','Day','Hour','District','Primary Type','Alarm']]
df1.head()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Assuming df1 is already defined and contains the necessary columns
X = df1[['Month', 'Day', 'Hour', 'District']]  # independent variables
y = df1['Alarm']  # dependent variable

# Split the dataset into training and testing sets (75% train, 25% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

# Initialize and fit the Decision Tree Classifier
d_tree = DecisionTreeClassifier(random_state=10)
d_tree = d_tree.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred = d_tree.predict(X_test)

# Model Evaluation
print("Accuracy:", metrics.accuracy_score(y_test, y_pred) * 100, "\n")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\n----------Confusion Matrix------------------------------------")
print(pd.DataFrame(cm, index=unique_labels(y_test), columns=unique_labels(y_pred)))

# Classification Report
print("\n----------Classification Report------------------------------------")
print(classification_report(y_test, y_pred))

# Unweighted Average Recall (UAR)
uar = 0
num_classes = len(unique_labels(y_test))
for i in range(num_classes):
    uar += cm[i][i] / cm[:, i].sum()
uar /= num_classes
print("\nUAR ->", uar)     #UAR stands for Unweighted Average Recall calculate the avg of each class of correct index and handles all possible classes. 
```
